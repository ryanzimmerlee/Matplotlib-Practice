```python
# Part 1: Macho Merging

# Import the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bring each CSV into a separate data frame
csv_path_2013 = (r"C:\Users\ryanz\Desktop\Data Analytics Bootcamp\Resources\RICH201901DATA3\05-Matplotlib\2\Activities\08-Stu_WinnerWrestling-Part1\Resources\WWE-Data-2013.csv")
csv_path_2014 = (r"C:\Users\ryanz\Desktop\Data Analytics Bootcamp\Resources\RICH201901DATA3\05-Matplotlib\2\Activities\08-Stu_WinnerWrestling-Part1\Resources\WWE-Data-2014.csv")
csv_path_2015 = (r"C:\Users\ryanz\Desktop\Data Analytics Bootcamp\Resources\RICH201901DATA3\05-Matplotlib\2\Activities\08-Stu_WinnerWrestling-Part1\Resources\WWE-Data-2015.csv")
csv_path_2016 = (r"C:\Users\ryanz\Desktop\Data Analytics Bootcamp\Resources\RICH201901DATA3\05-Matplotlib\2\Activities\08-Stu_WinnerWrestling-Part1\Resources\WWE-Data-2016.csv")

df_2013 = pd.read_csv(csv_path_2013)
df_2014 = pd.read_csv(csv_path_2014)
df_2015 = pd.read_csv(csv_path_2015)
df_2016 = pd.read_csv(csv_path_2016)

wrestler_df = pd.merge(df_2013, df_2014, how='outer', on='Wrestler')
wrestler_df

re_wrestler_df = wrestler_df.rename(columns={
    "Wins_x":"2013 Wins",
    "Losses_x":"2013 Losses",
    "Draws_x":"2013 Draws",
    "Wins_y":"2014 Wins",
    "Losses_y":"2014 Losses",
    "Draws_y":"2014 Draws"
})
re_wrestler_df.head()
```

```python
re2_wrestler_df = pd.merge(re_wrestler_df,df_2015,on='Wrestler',how='outer')
re2_wrestler_df

re2_wrestler_df = re2_wrestler_df.rename(columns={
    "Wins":"2015 Wins",
    "Losses":"2015 Losses",
    "Draws":"2015 Draws"
})
re2_wrestler_df.head()

```

```python
c_wrestler_df = pd.merge(re2_wrestler_df,df_2016,how='outer',on='Wrestler')
c_wrestler_df

clean_wrestler_df = c_wrestler_df.rename(columns={
    "Wins":"2016 Wins",
    "Losses":"2016 Losses",
    "Draws":"2016 Draws"
})
clean_wrestler_df.head()
```

```python
# Replace all NaN values with 0 
nonull_wrestler_df = clean_wrestler_df.fillna(0)
nonull_wrestler_df

# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values
nonull_wrestler_df['Total Wins'] = nonull_wrestler_df['2013 Wins'] + nonull_wrestler_df['2014 Wins'] + nonull_wrestler_df['2015 Wins'] + nonull_wrestler_df['2016 Wins']

# Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values
nonull_wrestler_df['Total Losses'] = nonull_wrestler_df['2013 Losses'] + nonull_wrestler_df['2014 Losses'] + nonull_wrestler_df['2015 Losses'] + nonull_wrestler_df['2016 Losses']

# Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values
nonull_wrestler_df['Total Draws'] = nonull_wrestler_df['2013 Draws'] + nonull_wrestler_df['2014 Draws'] + nonull_wrestler_df['2015 Draws'] + nonull_wrestler_df['2016 Draws']

# Create a new column called "Total Matches" and add up the total wins, losses, and draws for each wrestler to fill in the values
nonull_wrestler_df['Total Matches'] = nonull_wrestler_df['Total Wins'] + nonull_wrestler_df['Total Losses'] + nonull_wrestler_df['Total Draws']

nonull_wrestler_df
```

```python
# Create a new dataframe for those wrestlers who have wrestled at least 100 matches,
# have at least one win in 2013,
# and have at least one win in 2016
seasoned_wrestler = nonull_wrestler_df.loc[(nonull_wrestler_df['Total Matches'] >= 100) & (nonull_wrestler_df['2013 Wins'] > 0) & (nonull_wrestler_df['2016 Wins'] > 0)]
seasoned_wrestler

# Set the index of this new dataframe to be the wrestlers name
wr_final_df = seasoned_wrestler.set_index('Wrestler')
wr_final_df
```

```python
# Collect the user's input to search through our data frame
user_input = input("Please enter a wrestler's name to view their stats. " )
```

```python
# Create a series that looks for a wrestler by name and then traces their wins from 2013 to 2016
wins = wr_final_df.loc[user_input,["2013 Wins","2014 Wins", "2015 Wins", "2016 Wins"]]

# Create a series that looks for a wrestler by name and then traces their losses from 2013 to 2016
losses = wr_final_df.loc[user_input,["2013 Losses","2014 Losses","2015 Losses", "2016 Losses"]]
```

```python
# Create a list of the years that we will use as our x axis
years = [2013, 2014, 2015, 2016]

# Plot our line that will be used to track a wrestler's wins over the years
plt.plot(years, wins, color="green", label="Wrestler Wins")

# Plot our line that will be used to track a wrestler's losses over the years
plt.plot(years, losses, color="red", label="Wrestler Losses")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

plt.title(user_input + "'s Career Between 2013 - 2016")
plt.xlabel("Years")
plt.ylabel("Wins/Losses")

# Print our chart to the screen
plt.show()
```
